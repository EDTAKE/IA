```python
class State():
    
    def __init__(self,ncan,nmis,posbote):
        self.value = (ncan,nmis,posbote)
        self.ncaniz = ncan
        self.nmisiz = nmis
        self.posbote = posbote
    
    def __repr__(self):
        return '<Estado {}>'.format(self.value)
        
                                                      
    def __eq__(self, other):
        return isinstance(other, State) and self.value == other.value
    
    def __hash__(self):
        return hash(self.value)
    
    def getCanibalesIzq(self):
        return self.ncaniz
    
    def getCanibalesDer(self):
        return self.ncander
    
    def getMisionerosIzq(self):
        return self.nmisiz

    def getMisionerosDer(self):
        return self.nmisder
    
    def getPosBote(self):
        return self.posbote
    
    def setPosBote(self,posbote):
        self.posbote = posbote
    
    def setCanibales(self,ncaniz, ncander):
        self.ncaniz = ncaniz
        self.ncander = ncander
        
    def setMisioneros(self,nmisiz,nmisder):
        self.nmisiz = nmisiz
        self.nmisder = nmisder
            
        
```
